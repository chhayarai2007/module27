Q5. What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the
url_for() function.

ans: The url_for() function in Flask is used to generate URLs for a specific function. 
It accepts the name of the function as its first argument, and any number of keyword arguments, 
each corresponding to a variable part of the URL rule. 
For example, url_for('login', next='/') would generate a URL like /login?
